```python
#Yash Jain FE595 Assignment 1 
# Webscraping and NLP 

# Libraries to use 
import requests as re
import pandas as pd
from bs4 import BeautifulSoup
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Part 1 of the Assignment 
columns = ['Name','Purpose']
Data_with_50_companies = pd.DataFrame(columns=columns)
URL = 'http://3.85.131.173:8000/random_company'
asd = []
sdf = []
items_to_find = ['name', 'purpose']
for i in range(50):
    asd = []
    page = re.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    #print(soup)
    for item in items_to_find:
       element = str(soup.find_all('li', string=lambda text: item in text.lower()))
       asd.append(element)
    for i, elements in enumerate(asd):
        elements = elements.replace('<li>','')
        elements = elements.replace('</li>','')
        elements = elements.replace('[','')
        elements = elements.replace(']','')
        split_elements = elements.split(':')
        datalen = len(Data_with_50_companies)
        Data_with_50_companies.loc[datalen] = split_elements
        #Data_with_50_companies = Data_with_50_companies.T
    sdf.append(asd)
    #print(asd)
#print(sdf)

#print(Data_with_50_companies)
Data_with_50_companies = Data_with_50_companies.drop(['Name'],axis=1)
Data_with_50_companies
```

```python
Data_with_50_companies
df1= (Data_with_50_companies.iloc[::2].reset_index(drop=True))
df2 = (Data_with_50_companies.iloc[1::2].reset_index(drop=True))
df1.columns = ['Name']
df2.columns = ['Purpose']
df3 = pd.concat([df1,df2],axis=1)

df3.to_csv('Final_Excel.csv',index = False)
```
